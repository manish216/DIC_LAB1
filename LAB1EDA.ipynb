# Functions

```r
generateSquares <- function(x){ # define the function named generate squares
    for(i in 1:x){ 
        y <- i^2
        print(y)        
    }
    
    return(x^2) # return the values 
}

a = generateSquares(4) # Assiging the result of function to variable a
paste(c("The square value for given number is ", a))
```

```r
func_arguments <- function(a=1,b=2,c=3){ # define the function with default parameters
    res <- a+(b*c)
    print(res)
}
func_arguments() # calling the function with default parameters 
func_arguments(4,5,6) # calling the fucntions with parameters
func_arguments(a=5,b=6,c=7)
```

```r
sales1 <- c(12,14,16,29,30,45,19,20,16,19,34,20) # declaring the sales vector 
sales2 <- rpois(12,34) # declaring the poission distribution function with n = 12 (we want 12 values)
sales2                 #, and lambda =34 (mean for the distribution)
par(bg="cornsilk")  http://localhost:8888/notebooks/LAB1/LAB1EDA.ipynb#   # par is used to set any graphical parameter here we are background color has cornsilk
plot(sales1,col="blue",type ="o",ylim =c(0,100),xlab ="Month",ylab="Sales") # using plot function to plot sales1 vector 
title(main="Sales by Month") # sets the title for the plot

lines(sales2,type="o",pch=22,lty=2,col="red") # using line function to plot the graph for sales2 vector
grid(nx=NA,ny=NULL) # it is used to produce grid lines 
legend("topright",inset=0.05,c("sales1","sales2"),fill=c("blue","red"),horiz=TRUE)  
```

#### 3.  Below graph takes the same input as the one above but gives a different sales2 graph as we used rpois function which            gives random values.
